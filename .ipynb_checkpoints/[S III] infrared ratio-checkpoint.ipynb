```python
import numpy as np
import pyneb as pn
```

```python
s3 = pn.Atom('S', 3)
```

```python
s3.lineList
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s3.plotEmiss()
```

```python
densities = np.logspace(1.0, 5.0)
```

```python
em = s3.getEmissivity(1.e4, densities)
em.shape
```

```python
fig, ax = plt.subplots()
ax.plot(densities, em[0]/em[2])
```
